```python
frequency = df["Venue"].value_counts()
keys = frequency.keys().to_list()
counts = frequency.to_list()
plt.pie(x=counts , labels = keys , autopct = "%1.1f%%")
plt.title("Match Venue")
plt.show()
```

```python
frequency = df["Venue"].value_counts()
keys = frequency.keys().to_list()
counts = frequency.to_list()
my_list = [0.1,0.3,0.5,1,0.7,0.8]
plt.pie(x=counts , labels = keys , autopct = "%1.1f%%",explode = my_list)
plt.title("Match Venue")
plt.show()
```

```python
from matplotlib import cm
frequency = df["Venue"].value_counts()
keys = frequency.keys().to_list()
counts = frequency.to_list()
cs = cm.Set1([5,6])
plt.pie(x=counts , labels = keys , autopct = "%1.1f%%" , colors = cs)
plt.show()
```

```python
fre= df["Venue"].value_counts()
keys= fre.keys().to_list()
counts = fre.to_list()
mylist= [1,0.01]
plt.pie(x = counts,labels= keys , autopct = "%1.1f%%" )
circle = plt.Circle(xy = (0,0),radius = 0.4,color = "white")
plt.gcf()
plt.gca().add_artist(circle)
plt.title('Venue')
plt.show()
```
